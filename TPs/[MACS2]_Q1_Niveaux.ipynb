Avec ces listes, il sera possible de construire rapidement des sous-tables comme `df[Info]` par exemple.

````python
Info = df.columns.str.contains("Info")
Info = df.columns[Info].tolist()
Maths = df.columns.str.contains("Maths")
Maths = df.columns[Maths].tolist()
Toolboxes = df.columns.str.contains("Toolboxes")
Toolboxes = df.columns[Toolboxes].tolist()
print(df[Info])
````

### 4.3 Affichage d'un boxplot des moyennes par domaine.

1. Commencez par construire une table à deux colonnes : Note et Domaine dans la quelle la note correspond à la moyenne d'un participant de le domaine (Info, Toolbox ou Maths). On aura 3 fois plus de lignes que la table initiale (le nom du participant n'est pas utile).
2. Utilisez 'seaborn' pour afficher les trois boites à moustaches.

````python
di = pd.DataFrame(df[Info].mean(axis=1), columns=['Note'])
di['Domaine'] = 'Info'

dm = pd.DataFrame(df[Maths].mean(axis=1), columns=['Note'])
dm['Domaine'] = 'Maths'

dt = pd.DataFrame(df[Toolboxes].mean(axis=1), columns=['Note'])
dt['Domaine'] = 'Toolboxes'

df_moyenne = pd.concat([di, dm, dt], axis=0)
# df_moyenne

sns.boxplot(data=df_moyenne, x='Domaine', y = 'Note', orient='v')
plt.title('Moyenne des compétences')
````

### 4.4 Affichage des trois moyennes par participant.

Aide : recombinez les données précédentes dans l'autre sens (horizontalement au lieu de verticalement), rajoutez l'index correspondant au nom et utilisez `sns.heatmap(...)`.

````python
db = pd.DataFrame(dict(Info=di['Note'].values, Toolboxes=dt['Note'].values, Maths=dm['Note'].values), index=df['Nom d\'utilisateur'])

sns.heatmap(data=db, fmt='.1f', annot=True)
````
